```python
%matplotlib inline

import numpy as np
import pandas as pd

import nested_plot as nplot

import seaborn as sns
sns.set()
```

```python
image_df = pd.read_csv('data_catalogues/processing_2018-03-08_14-20-00/120318_reclass_and_base.csv', index_col=0)
image_df.head()
```

```python
interior_rooms = ['bathroom',
                  'bedroom',
                  'conservatory', 
                  'diningroom',
                  'empty',
                  'entrance',
                  'graphic', 
                  'kitchen',
                  'livingroom', 
                  'misc_int', 
                  'study']

data = image_df.loc[(image_df['room_reclass'].isin(interior_rooms))]
data.sample(4)
```

```python
label_fn = nplot.label_before_after
label_info = {'before': 'room_base',
              'after': 'room_reclass'}


nplot.image_grid(data, image_path='image_path_reclass',  image_size='xl', save=True,
           fig_title='Reclassified images', fill_grid=True, f_stem='TEST_',
          label_info=label_info, label_fn=label_fn, annotate=True)
               # output_dir = './image_grids/')
```
